```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Data Collection and Processing
```

```python
#loading the CSV data to our pandas data frames
heart_data = pd.read_csv('/content/drive/MyDrive/Heart Disease Predictor/heart_disease_data.csv')
```

```python
#checking the read file by printing the first five rows
heart_data.head(5)
```

```python
heart_data.tail(5)
```

```python
#check the number of rows and columns in the data set
heart_data.shape
```

```python
#getting info about the data
heart_data.info()
```

```python
#checking for missing values
heart_data.isnull().sum()
```

```python
#statistical measure about the data
heart_data.describe()
```

```python
#counting the number of people with heart disease
heart_data['target'].value_counts()
```

```python
#Splitting the features and target
```

```python
x=heart_data.drop(columns='target', axis=1)
y=heart_data['target']
```

```python
print(x)
```

```python
print(y)
```

```python
#Splitting the data into training data and test data
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, stratify=y, random_state=3)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#Model Evaluation
```

```python
#Accuracy Score
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy: ", training_data_accuracy)
```

```python
#Accuracy Score for test data
x_test_prediction=model.predict(x_test)
training_data_accuracy1=accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy of test data: ", training_data_accuracy1)
```
